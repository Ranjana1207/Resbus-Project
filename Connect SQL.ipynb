```python
#import Libraries
import pandas as pd
import mysql.connector
import numpy as np
```

```python
#all csv change to dataframe
df_buses_1=pd.read_csv("df_buses_1.csv")
df_buses_2=pd.read_csv("df_buses_2.csv")
df_buses_3=pd.read_csv("df_buses_3.csv")
df_buses_4=pd.read_csv("df_buses_4.csv")
df_buses_5=pd.read_csv("df_buses_5.csv")
df_buses_6=pd.read_csv("df_buses_6.csv")
df_buses_7=pd.read_csv("df_buses_7.csv")
df_buses_8=pd.read_csv("df_buses_8.csv")
df_buses_9=pd.read_csv("df_buses_9.csv")
df_buses_10=pd.read_csv("df_buses_10.csv")

Final_df=pd.concat([df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,
                    df_buses_6,df_buses_7,df_buses_8,df_buses_9,df_buses_10],ignore_index=True)
Final_df
```

```python
#DATA CLEANING

#Data about data
Final_df.info()
```

```python
# Ensure that the 'Prices' column is of string type
Final_df["Prices"] = Final_df["Prices"].astype(str)

# Remove the "INR" text from the 'Prices' column
Final_df["Prices"] = Final_df["Prices"].str.replace("INR", "", regex=False)

# Strip any leading or trailing whitespace (if necessary)
Final_df["Prices"] = Final_df["Prices"].str.strip()

# Convert the cleaned prices to numeric format (float)
Final_df["Prices"] = pd.to_numeric(Final_df["Prices"], errors='coerce')

# Fill NaN values with 0 (optional)
Final_df["Prices"].fillna(0, inplace=True)

# Final DataFrame
print(Final_df.head())
```

```python
Final_df["Prices"]
```

```python
import pandas as pd

# Ensure the 'Ratings' column is of string type
Final_df["Ratings"] = Final_df["Ratings"].astype(str)

# Remove the word "New" from the 'Ratings' column
Final_df["Ratings"] = Final_df["Ratings"].str.replace("New", "", regex=False)

# Strip any leading or trailing whitespace
Final_df["Ratings"] = Final_df["Ratings"].str.strip()

# Split the string and take the first part (assuming ratings are like "4.5 out of 5")
Final_df["Ratings"] = Final_df["Ratings"].str.split().str[0]

# Convert the cleaned ratings to numeric format (float)
Final_df["Ratings"] = pd.to_numeric(Final_df["Ratings"], errors='coerce')

# Fill NaN values with 0
Final_df["Ratings"].fillna(0, inplace=True)

# Final DataFrame
print(Final_df.head())
```

```python
#info after the data type change
Final_df.info()
```

```python
Final_df=Final_df[Final_df["Prices"] <= 7000]
```

```python
#replace nan-value
Final_df=Final_df.replace({ np.nan:None})
```
